# Route Consolidation
Combines multiple route files

```python
import xml.etree.ElementTree as ET
```

```python

files = ["../data_basis/cicero_mueller_routes.rou.xml", "../preprocessing_output/deadheads_cicero_mueller.rou.xml"]
root = ET.Element("routes")

for file in files:
    tree = ET.parse(file)
    for route in tree.getroot():
        root.append(route)

ET.ElementTree(root).write(
    "../preprocessing_output/deadheads_routes_cicero_mueller.rou.xml",
    encoding="UTF-8",
    xml_declaration=True,
)
```
