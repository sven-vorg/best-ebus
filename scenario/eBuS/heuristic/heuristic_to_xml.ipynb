```python
# Imports
import json
import xml.etree.ElementTree as ET
from xml.dom import minidom
import pandas as pd
```

```python
# Helper Function
ADDITIONAL_TIMES = "../route_building/trips_Cicerostraße.txt"
df = pd.read_csv(ADDITIONAL_TIMES,sep=";")
# ADDITIONAL_EDGES = "../route_building/cicero_mueller_routes.rou.xml"

def trip_to_route(trip_id: int) -> tuple[str, float, float]:
    row = df.loc[df["TRIP_ID"] == trip_id].iloc[0]

    return (
        row["ORIGINAL_TRIP_ID"],
        row["START_STOP_ID"],
        row["END_STOP_ID"],
    )

def trip_list_to_route(trip_list:[int]) -> [tuple[str, float, float]]:
    route = []
    for trip in trip_list:
        route.append(trip_to_route(trip))
    print(route)
    return route
```
